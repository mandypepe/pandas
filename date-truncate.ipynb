```python
import pandas as pd
import datetime as dt
import pandas_datareader as pdr
company= "BAC"
start="2014-01-01"
end="2018-12-31"
df=pdr.DataReader(name=company,data_source='iex',start=start,end=end)
df.index=pd.to_datetime(df.index)
df.info()
df.sort_index(inplace=True)
```
